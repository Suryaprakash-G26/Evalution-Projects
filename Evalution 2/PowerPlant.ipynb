```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
url1='https://raw.githubusercontent.com/FlipRoboTechnologies/ML_-Datasets/main/Global%20Power%20Plant%20Database/database_AUS.csv'
url2='https://raw.githubusercontent.com/FlipRoboTechnologies/ML_-Datasets/main/Global%20Power%20Plant%20Database/database_IND.csv'
url3='https://raw.githubusercontent.com/FlipRoboTechnologies/ML_-Datasets/main/Global%20Power%20Plant%20Database/database_USA.csv'
df1=pd.read_csv(url1)
df2=pd.read_csv(url2)
df3=pd.read_csv(url3)
df = pd.concat([df1, df2, df3], ignore_index=True)
df
```

```python
df = df.drop(['estimated_generation_gwh', 'wepp_id', 'country_long','gppd_idnr','latitude','longitude',
              'generation_data_source','year_of_capacity_data','geolocation_source','source','url','owner'
            ], axis=1)
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['name'].value_counts()
```

```python
df.drop(['name'],axis=1,inplace=True)
```

```python
non_numeric = df[~df['generation_gwh_2019'].apply(lambda x: str(x).replace('.', '', 1).isdigit())]
print(non_numeric['generation_gwh_2019'])
```

```python
df['generation_gwh_2019'] = pd.to_numeric(df['generation_gwh_2019'], errors='coerce')
```

```python
df['commissioning_year']=df['commissioning_year'].fillna(df['commissioning_year'].median())
df['generation_gwh_2013']=df['generation_gwh_2013'].fillna(df['generation_gwh_2013'].median())
df['generation_gwh_2014']=df['generation_gwh_2014'].fillna(df['generation_gwh_2014'].median())
df['generation_gwh_2015']=df['generation_gwh_2015'].fillna(df['generation_gwh_2015'].median())
df['generation_gwh_2016']=df['generation_gwh_2016'].fillna(df['generation_gwh_2016'].median())
df['generation_gwh_2017']=df['generation_gwh_2017'].fillna(df['generation_gwh_2017'].median())
df['generation_gwh_2018']=df['generation_gwh_2018'].fillna(df['generation_gwh_2018'].median())
df['generation_gwh_2019']=df['generation_gwh_2019'].fillna(df['generation_gwh_2019'].median())
```

```python
df.info()
```

```python
df = df.drop(['other_fuel2', 'other_fuel3'], axis=1)
df.head()
```

```python
df['other_fuel1']=df['other_fuel1'].fillna(df['other_fuel1'].mode()[0])
```

```python
df.info()
```

```python
df['capacity_mw']=df['capacity_mw'].fillna(df['capacity_mw'].median())
```

```python
from sklearn.ensemble import RandomForestClassifier  
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
train_df = df[df['primary_fuel'].notna()]
test_df = df[df['primary_fuel'].isna()]
```

```python
X_train = train_df.drop(columns=['primary_fuel'])
y_train = train_df['primary_fuel']
X_test = test_df.drop(columns=['primary_fuel'])
```

```python
categorical_cols = X_train.select_dtypes(include=['object']).columns
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols)
    ],
    remainder='passthrough'
)
```

```python
model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier())
])
```

```python
model.fit(X_train, y_train)
```

```python
predicted_values = model.predict(X_test)
df.loc[df['primary_fuel'].isna(), 'primary_fuel'] = predicted_values
```

```python
maxAccu=0
maxRS=0
for i in range(0,100):
    x_train,x_test,y_train,y_test = train_test_split(x_scaler,y,test_size=.30, random_state=i)
    lr=LinearRegression()
    lr.fit(x_train, y_train)
    pred = lr.predict(x_test)
    acc=r2_score(y_test, pred)
    if acc>maxAccu:
        maxAccu=acc
        maxRS=i
print('Maximum r2 score is',maxAccu,'on Random_state',maxRS)
```
